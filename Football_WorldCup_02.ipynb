```python
import pandas as pd
import matplotlib.pyplot as plt
Mydata3=pd.read_csv('WorldCups.csv',delimiter=",")
Mydata3
```

```python
import matplotlib.pyplot as plt
from numpy.random import normal,rand
y= Mydata3.loc[:,'GoalsScored']
x=Mydata3.loc[:,'Winner']
z= Mydata3.loc[:,'Year']
plt.ylabel('GoalsScored')
plt.title('Years')
plt.plot(z,y,'ro',z,y,'b-')
```

```python
colors = Mydata3.loc[:,'GoalsScored']
plt.xlabel('Years')
plt.ylabel('Goal Scored')
plt.title('Scattered plot of goal scored countries ')
plt.scatter(x, y,c=colors, alpha=0.1)
```

```python
Bra=Arg=Ger=Urg=Fra=Eng=Ita=Spa=0
for i in range(0, len(Mydata3)):
    if Mydata3.loc[i]['Winner']=='Brazil':
        Bra+=1
    if Mydata3.loc[i]['Winner']=='Germany FR':
        Ger+=1
    if Mydata3.loc[i]['Winner']=='Germany':
        Ger+=1
    if Mydata3.loc[i]['Winner']=='Uruguay':
        Urg+=1
    if Mydata3.loc[i]['Winner']=='Italy':
        Ita+=1
    if Mydata3.loc[i]['Winner']=='England':
        Eng+=1
    if Mydata3.loc[i]['Winner']=='France':
        Fra+=1
    if Mydata3.loc[i]['Winner']=='Argentina':
        Arg+=1
    if Mydata3.loc[i]['Winner']=='Spain':
        Spa+=1
raw_data = {'name':['Brazil', 'Germany','Uruguay','Italy','England','France','Argentina','Spain'],
		        'Championship Number':[Bra, Ger,Urg,Ita,Eng,Fra,Arg,Spa],
                
		        }
df = pd.DataFrame(raw_data, columns = ['name','Championship Number'])
print(df)
```

```python
Name=['Brazil', 'Germany','Uruguay','Italy','England','France','Argentina','Spain']
Win=[Bra, Ger,Urg,Ita,Eng,Fra,Arg,Spa]
plt.ylabel('Win')
plt.title('Championship')
plt.plot(Name,Win,'ro')
```

```python
labels = 'Brazil', 'Germany','Uruguay','Italy','England','France','Argentina','Spain'
sizes =df.loc[:,'Championship Number']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','salmon','darkorange','peachpuff','lightgreen']
         
explode = (0.1, 0.1, 0, 0.1,0,0,0,0)  # explode 1st slice
 
# Plot

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.title('Championship of Teams in World Cup')
plt.show()
```
