```python
import numpy as np
import pandas as pd
Mydata=pd.read_csv('WorldCupMatches.csv',delimiter=",")
Mydata.shape
data=Mydata.loc[:851,['Year','Stadium','Home Team Name','Home Team Goals','Away Team Goals','Away Team Name','Win conditions']]

for i in range(0, len(data)):
    if data.loc[i]['Home Team Goals']>data.loc[i]['Away Team Goals']:
        data.loc[i,'Win conditions']=data.loc[i,'Home Team Name']
    elif data.loc[i]['Home Team Goals']<data.loc[i]['Away Team Goals']:
        data.loc[i,'Win conditions']=data.loc[i,'Away Team Name']
    else:
        data.loc[i,'Win conditions']='equal'
```

```python
#Brazil vs Argentina
A1=A2=Aq=Ap=0
for i in range(0, len(data)):
    if (data.loc[i,'Home Team Name']=='Argentina')& (data.loc[i,'Away Team Name']=='Brazil'):
        Ap+=1
        if data.loc[i,'Win conditions']=='Brazil':
            A1+=1
        elif data.loc[i,'Win conditions']=='equal':
            Aq+=1
    elif (data.loc[i,'Home Team Name']=='Brazil')& (data.loc[i,'Away Team Name']=='Argentina'):
        Ap+=1
        if data.loc[i,'Win conditions']=='Argentina':
            A2+=1
        elif data.loc[i,'Win conditions']=='equal':
            Aq+=1
```

```python
#Brazil vs Germany
B1=B2=Bq=Bp=0
for i in range(0, len(data)):
    if (data.loc[i,'Home Team Name']=='Germany')& (data.loc[i,'Away Team Name']=='Brazil'):
        Bp+=1
        if data.loc[i,'Win conditions']=='Brazil':
            B1+=1
        elif data.loc[i,'Win conditions']=='equal':
            Bq+=1
    elif (data.loc[i,'Home Team Name']=='Brazil')& (data.loc[i,'Away Team Name']=='Germany'):
        Bp+=1
        if data.loc[i,'Win conditions']=='Germany':
            B2+=1
        elif data.loc[i,'Win conditions']=='equal':
            Bq+=1
for j in range(0, len(data)):
    if (data.loc[j,'Home Team Name']=='Germany FR')& (data.loc[j,'Away Team Name']=='Brazil'):
        Bp+=1
        if data.loc[j,'Win conditions']=='Brazil':
            B1+=1
        elif data.loc[j,'Win conditions']=='equal':
            Bq+=1
    elif (data.loc[j,'Home Team Name']=='Brazil')& (data.loc[i,'Away Team Name']=='Germany FR'):
        Bp+=1
        if data.loc[j,'Win conditions']=='Germany FR':
            B2+=1
        elif data.loc[j,'Win conditions']=='equal':
            Bq+=1
```

```python

ww=['Austria','Hungary','Uruguay','Italy','England','France','Sweden','Switzerland','Turkey','Yugoslavia','Argentina','USA','Spain','Portugal']
hh=[]
for h in ww:
    C1=C2=Cq=Cp=0
    for i in range(0, len(data)):
        if ((data.loc[i,'Home Team Name']==h)& (data.loc[i,'Away Team Name']=='Brazil'))&(data.loc[i,'Win conditions']=='Brazil'):
            Cp+=1
            C1+=1
        elif ((data.loc[i,'Home Team Name']==h)& (data.loc[i,'Away Team Name']=='Brazil'))&(data.loc[i,'Win conditions']==h):
            C2+=1
            Cp+=1
        elif ((data.loc[i,'Home Team Name']==h)& (data.loc[i,'Away Team Name']=='Brazil'))&(data.loc[i,'Win conditions']=='equal'):
            Cq+=1
            Cp+=1
        elif ((data.loc[i,'Home Team Name']=='Brazil')& (data.loc[i,'Away Team Name']=='Italy'))&(data.loc[i,'Win conditions']=='Brazil'):
            Cp+=1
            C1+=1 
        elif ((data.loc[i,'Home Team Name']=='Brazil')& (data.loc[i,'Away Team Name']=='Italy'))&(data.loc[i,'Win conditions']==h):
            Cp+=1
            C2+=1 
        elif ((data.loc[i,'Home Team Name']=='Brazil')& (data.loc[i,'Away Team Name']=='Italy'))&(data.loc[i,'Win conditions']=='equal'):
            Cq+=1
            Cp+=1 
        ht=[C1,C2,Cq,Cp]
    hh.append(ht)
```
