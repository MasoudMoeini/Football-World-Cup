```python
import numpy as np
import pandas as pd
Mydata=pd.read_csv('WorldCupMatches.csv',delimiter=",")
Mydata.shape
data=Mydata.loc[:851,['Year','Stadium','Home Team Name','Home Team Goals','Away Team Goals','Away Team Name','Win conditions']]

for i in range(0, len(data)):
    if data.loc[i]['Home Team Goals']>data.loc[i]['Away Team Goals']:
        data.loc[i,'Win conditions']=data.loc[i,'Home Team Name']
    elif data.loc[i]['Home Team Goals']<data.loc[i]['Away Team Goals']:
        data.loc[i,'Win conditions']=data.loc[i,'Away Team Name']
    else:
        data.loc[i,'Win conditions']='equal'
```
